1. (5 Points) Write a Python/NumPy code block that finds the distinct/unique common items between these two arrays:
a = np.array([1, 2, 3, 2, 3, 4, 3, 4, 5, 6])
b = np.array([7, 2, 10, 2, 7, 4, 9, 4, 9, 8])
Your output should contain only the distinct overlapping values. For example, if a 2 is found in both array a and array b,
your output should contain only one 2 even if array a contains more than one 2 within it.


Answer 1: Find the unique values inside of both a and b.
Iterate over the unique values in a and ensure that if there is a mathcing value in b it is printed.

```
# import the numpy library under the alias np
import numpy as np
#Create the two numpy arrays a and b.
a = np.array([1, 2, 3, 2, 3, 4, 3, 4, 5, 6])
b = np.array([7, 2, 10, 2, 7, 4, 9, 4, 9, 8])
#Find all unique values in the numpy array a and b. 
#Then assign these unique values to u_a and u_b respectivley.
u_a=np.unique(a)
u_b=np.unique(b)
#Iterate over the unique values in a(u_a).
for i in u_a:
# Make sure that only values in u_a that match value in u_b are printed 
  if i in u_b:
    print(i)
```

3. (5 Points) You are given the following two arrays:
a = np.array([12, 5, 7, 15, 3, 1, 8])
b = np.array([14, 6, 3, 11, 19, 12, 5])
Write a Python/Numpy code block that removes from array a any items that are also present in array b.


Answer 3: Create the two numpy arrays a and b. Then find the unique values in both a and b.
Iterate over the unique values in a and make to print them if none of the values in a are found in b.

```
#Create the two numpy arrays a and b.
a = np.array([12, 5, 7, 15, 3, 1, 8]) 
b = np.array([14, 6, 3, 11, 19, 12, 5])
#Find all unique values in the numpy array a and b. 
#Then assign these unique values to u_a and u_b respectivley.
u_a=np.unique(a)
u_b=np.unique(b)
#Iterate over the unique values in a(u_a).
for i in u_a:
# Make sure that only values that are not in u_b are printed
  if i not in u_b:
    print(i)
```

Set up to answer question 4: Create the array miles that has 35 entries that add up to 1519, which is the distance from New York to key west Florida.Four of the entries are left blank to simulate rain or otherwise a day spent not cycling. Calculate the cumalive sum of miles to generate your odometer data.

```
#Create the array miles that has 35 entries that add up to 1519
miles = np.array([43,47,35,46,44,52,37,42,39,29,45,0,46,57,58,62,49,51,48,52,0,57,46,53,45,61,54,43,43,54,0,0,43,75,63])
#Print the sum of miles to make sure that it does indeed match 1519
print (np.sum(miles))
#Calculate the cumalive sum of miles to generate your odometer data.
cum_sum=np.cumsum(miles)
#Print the cumsum of miles to make sure that we have our array of odmeter data by day.
print(cum_sum)
```

Answer 4: Your odometer data is saved in cum_sum from the last code cell. Turn it into a list then insert a 0 before the first data point. Turn the data back into a np.array and save it as new_c_s(new Cum Sum).
Use numpys built in diff function to determine how many miles were biked on each individual day.
Note: it is nessecary to insert the 0 in to the list at index 0 or you will be missing your first days cycling data.

```
#You can easily pass cum_sum into the diff function to return the amount cycled each day.
#The only problem is diff would not return the first days cycling data. 
#We solve this by adding a zero to the array before the first value so that diff can fail to return it instead of our first days cycling data.

#Turn cum_sum into a list
list_c_s=list(cum_sum)
#Insert a zero at index zero into the list.
list_c_s.insert(0,0)
#Turn the list back into a Numpy array and save it as new_c_s
new_c_s=np.array(list_c_s)
#Run the diff function on new_c_s and print it.
original=np.diff(new_c_s)
print(original)
```
